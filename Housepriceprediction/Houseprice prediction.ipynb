```python
columns_to_remove=['date','yr_renovated','street','city','statezip', 'country']
data=data.drop(columns=columns_to_remove)
```

```python
data.head()
```

```python
import scipy.stats as stats
numeric_data = data.select_dtypes(include=['float64', 'int64'])
z_scores = stats.zscore(numeric_data)
threshold=1
print('size before removing outliers:',data.shape)
outliers_data=data[(z_scores>threshold).any(axis=1)]
data=data[(z_scores<=threshold).all(axis=1)]
print('After removing outliers:',data.shape)
```

```python
outliers_data.head()
```

```python
#data preprocessing
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data)
data_scale=pd.DataFrame(scaler.transform(numeric_data),columns=numeric_data.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
X=data.drop('price',axis=1)
y=data['price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
print(f"'xtrain{X_train.shape}")
print(f"'xtest{X_test.shape}")
```

```python
models=[('Random Forest',RandomForestRegressor()),
       ('Linear Regressor',LinearRegression()),
       ('Decision Tree',DecisionTreeRegressor()),
        ('KNN',KNeighborsRegressor())]
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
print(X_train.info())
print(X_test.info())
```

```python
for name,model in models:
    print(name)
    print()
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    print('Mean_Squared_Error:',mean_squared_error(y_test,y_pred))
    print('R2_score:',r2_score(y_test,y_pred))
    print('Mean_Absolute_error:',mean_absolute_error(y_test,y_pred))
```
